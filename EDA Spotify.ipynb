```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Check Null Values

pd.isnull(df_tracks).sum()
```

```python
# Check data types and count

df_tracks.info()
```

```python
# Value Count of Popularity scale

df_tracks.popularity.value_counts()
```

### 10 Least Popular Songs

```python
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

### 10 Most Popular Songs and greater than 90

```python
most_popular = df_tracks.query('popularity>90').sort_values('popularity', ascending=False).head(10)
most_popular
```

```python
# Set Index (Change Index)

df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks
```

```python
# Find Artist at particular location
# loc uses label/name and iloc uses position/index

df_tracks[['artists']].iloc[5]
```

```python
# Convert duration ms to s

df_tracks['duration'] = round(df_tracks['duration_ms']/1000)
df_tracks
```

```python
# Correlation between spotify feautures using heatmap

# Create matrix data 

tracks = df_tracks.corr()
tracks
```

```python
sn.set_context('paper')
plt.figure(figsize=(15,7))
sn.heatmap(data=tracks, cmap='Blues', annot=True, linewidth=1, linecolor='black').set(title='Correlation Heatmap between variable')
```

```python
## Create sample data from given data

df_sample = df_tracks.sample(int(0.004*len(df_tracks)))
len(df_sample)
```

```python
# Regression Plot

plt.figure(figsize=(10,4))
sn.regplot(data=df_sample, x='loudness', y='energy',).set_title("Loudness vs Engergy Correlation")
```

```python
# Regression Plot

plt.figure(figsize=(10,4))
sn.regplot(data=df_sample, x='popularity', y='acousticness',).set_title("Loudness vs Engergy Correlation")
```

```python
# Pull out only
year = df_tracks.index.year
year
```

```python
# Histogram for Year and No of songs

sn.displot(year, discrete=True, aspect= 2).set(title="Number of songs per year")
```

```python
# Barplot for relation between year and duration
duration = df_tracks['duration']

plt.figure(figsize=(20,10))
sn.barplot(x=year, y=duration, data=df_sample)
plt.xticks(rotation=90)
```

```python
# Lineplot fro relation between year and duration
# Index - release_date had duplicate values, so to remove any error 
plt.figure(figsize=(20,10))
sn.set_context('paper', font_scale=2)
df_sample['year'] = df_sample.index.year
sn.lineplot(data=df_sample.reset_index(), x='year', y='duration').set(title='Year v/s Duration')
```

```python
#Distribution plot (KDE) 

# left skew (peak high on right) - mean is less than median
# right skew (peak high on left) - mean is greater than median

df_sample=df_sample.reset_index()
sn.kdeplot(df_sample['duration'])
```

```python
# Box plot showing outliers for year vs duration

sn.set_context('paper', font_scale=1.5)
plt.figure(figsize=(20,50))
sn.boxplot(y='year', x='duration',orient="h",data=df_sample.reset_index())
plt.xticks(rotation=90)
```

```python
# Removing max outliers using clip method

lower = df_sample.duration.quantile(.05)
upper = df_sample.duration.quantile(.95)

sn.set_context('paper', font_scale=1.5)
plt.figure(figsize=(20,50))
df_sample['duration'] = df_sample.duration.clip(lower,upper)

sn.boxplot(y='year', x='duration',orient="h", data=df_sample)
```
